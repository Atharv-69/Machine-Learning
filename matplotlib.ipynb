```python
import matplotlib.pyplot as plt

x = ["python","c","c++","java"]
y=[90,80,75,85]
color=["r","g","y","orange"]
size=[100,200,300,400]
plt.scatter(x,y,c=color,s=size,alpha=0.99,marker="*",edgecolor="black",linewidth=1)
t=plt.colorbar()
t.title("Color Bar")
plt.title("Scatter plot",fontsize=18)
plt.xlabel("Programming language",fontsize=15)
plt.ylabel("Popularity ",fontsize=15)
plt.show() 
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = ["python","c","c++","java"]
y=[90,80,75,85]
color=["r"]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("popularity graph")
plt.xlabel("lang")
plt.ylabel("percentage")
plt.bar(p,y,width,color=color,align="edge",edgecolor="black",linewidth=3,alpha=0.6,label="popu")
plt.bar(p1,z,width,color="b",align="edge",edgecolor="black",linewidth=3,alpha=0.6,label="popu1")

plt.xticks(p+width,x,rotation=15)

plt.legend()
plt.show()
```
