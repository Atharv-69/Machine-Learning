```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.stats as stats 
import seaborn as sns
```

```python
df = pd.read_csv("D:\\air_quality_dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(['City','Country','Date','PM2.5','PM10'],axis=1)
y = df['PM10'] 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
plt.figure(figsize=(12,2))
plt.subplot(121)
sns.distplot(x_train['NO2'])
plt.show()

plt.subplot(122)
sns.distplot(x_train['SO2'])
```

```python
plt.figure(figsize=(12,2))
plt.subplot(121)
sns.distplot(x_train['CO'])
plt.show()

plt.subplot(122)
sns.distplot(x_train['O3'])
```

```python
plt.figure(figsize=(12,2))
plt.subplot(121)
sns.distplot(x_train['Temperature'])
plt.show()

plt.subplot(122)
sns.distplot(x_train['Wind Speed'])
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
regr = AdaBoostRegressor(random_state=0 , n_estimators=10)
```

```python
regr.fit(x_train,y_train)
```

```python
y_pred = regr.predict(x_test)
```

```python
regr.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score , r2_score
```

```python
r2_score  (y_test , y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
r2_score(y_test,y_pred_rf)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
r2_score(y_test , y_pred_dt)
```
